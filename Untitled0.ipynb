```python
# programa que peça as 4 notas bimestrais e mostre a média.

nota1 = float(input('insira a nota do 1° bimestre '))
nota2 = float(input('insira a nota do 2° bimestre '))
nota3 = float(input('insira a nota do 3° bimestre '))
nota4 = float(input('insira a nota do 4° bimestre '))

soma_notas=(nota1 + nota2 + nota3 + nota4)
media=(soma_notas/4)
print('A média do ano letivo foi ', media)
```

```python
# Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.

salario = float(input('Qual seu salário mensal? '))
hora = float(input('Quantas horas você trabalha por semana? '))

salario_semanal = salario/4
salario_hora = (salario_semanal/hora)
print('seu salário por hora é ', salario_hora)
```
